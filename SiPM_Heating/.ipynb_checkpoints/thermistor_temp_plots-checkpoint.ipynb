```python
#Import Statements
import numpy as np
import matplotlib.pyplot as plt
import csv
import datetime
import math
from scipy import optimize
```

```python
#Path of CSV Dataset
path_name = 'CSVDatasets/210819180000_210820104400.csv'

#starting and ending time variables for graphing (in minutes)
time1 = 900
time2 = len(time_u)-1

#starting and ending time variables for average temperature calculation (in minutes)
start_avg = 900
end_avg = 960
```

```python
#Arrays for thermistor data
t0,t1,t2,t3,t4,t5,t6,t7 = [],[],[],[],[],[],[],[]

#Time array
time_u = []

#Ambient temp array
amb_temp = []
```

```python
#Converts ADC values to celsius for thermistors 2-7
def cal(R):
    R = adc2R(float(R))
    beta=3892.0
    T0=25.0+273.15
    R0=10000
    r_inf = R0*math.exp(-beta/T0)
    T = beta/math.log(R/r_inf)
    return T - 273.15

#Converts ADC values to celsius for thermistors 0 and 1
def cal_surface(R):
    R = adc2R(float(R))
    beta=3940
    T0=25.0+273.15
    R0=10000
    r_inf = R0*math.exp(-beta/T0)
    T = beta/math.log(R/r_inf)
    return T - 273.15

#Helper function for both cal conversions
def adc2R(c, R=19e3, Vref=2.5):
    V = c*Vref/4096
    return V*R/(Vref-V);

#Fixes data values in csv file using err and cal
with open(path_name) as csvfile: 
    data = csv.reader(csvfile)
    for i, row in enumerate(data):
        t0.append(cal_surface(row[1]))
        t1.append(cal_surface(row[2]))
        t2.append(cal(row[3]))
        t3.append(cal(row[4]))
        t4.append(cal(row[5]))
        t5.append(cal(row[6]))
        t6.append(cal(row[7]))
        t7.append(cal(row[8]))
        amb_temp.append(float(row[10]))
        time_u.append(row[13])
```

```python
first_time = datetime.datetime.strptime(time_u[0],'%Y-%m-%d %H:%M:%S')

time_s = []
for i in time_u: #converts the hours time stamp to time elapsed
    elapsed = datetime.datetime.strptime(i,'%Y-%m-%d %H:%M:%S') - first_time
    time_s.append((elapsed.total_seconds())/60)

#Allows splicing of the plot at specific timestamps
def get_timestamp(minutes):
    for i in time_s:
        if i > minutes:
            return time_s.index(i) - 1
```

```python
#plots non-normalized thermistor readout values
def raw_plot(tx,c,n):
    #[get_timestamp(280):get_timestamp(340)]
    plt.plot(time_s[get_timestamp(time1):get_timestamp(time2)], np.asarray(tx)[get_timestamp(time1):get_timestamp(time2)], color = c, label = 'Thermistor '+n,  linestyle = 'none', marker = '.', markersize = 2)

#Plots the running average of a thermistor dataset
def mean_plot(tx,c,n):
    total_sum = 0
    mean_tx = []
    for x in enumerate(tx):
        total_sum += x[1]
        mean = total_sum / (x[0]+1)
        mean_tx.append(mean)
    plt.plot(time_s[get_timestamp(time1):get_timestamp(time2)], np.asarray(mean_tx)[get_timestamp(time1):get_timestamp(time2)], color = c, label = 'Thermistor '+n,  linestyle = 'none', marker = '.', markersize = 2)
```

```python
#plots the non-normalized thermistor values
#raw_plot(t0, 'g', '0') #On module
raw_plot(t1, 'r', '1') #Side 1 Module
raw_plot(t2, 'deeppink', '2') #Housing Side Far
raw_plot(t3, 'yellow', '3') #Housing Side Near
raw_plot(t4, 'orange', '4') #Cooling plate
raw_plot(t5, 'purple', '5') #Housing bottom
raw_plot(t6, 'aqua', '6') #Aluminium plate, centered
raw_plot(t7, 'blue', '7') #Aluminum plate, far right

plt.xlabel('Time / min', fontsize =16)
plt.ylabel('Thermistor temperature / $^\circ$C', fontsize =16)

plt.legend(prop={'size':14}, bbox_to_anchor=(1., 1), markerscale=5)
plt.title("Steady-state Readings at Room Temperature")
plt.grid()
plt.show()
```

```python
#thermistor normalization shifts: determined by hand from steady-state data
shifts = {'0': -0.25, '1': -0.254, '2': 0, '3': -0.014, '4': -0.027, '5': 0.029, '6': 0.118, '7': 0.196}    

#plots thermistor readout values normalized using the dictionary of shifts
def specific_norm(tx, c, n):
    plt.plot(time_s[get_timestamp(time1):get_timestamp(time2)], np.asarray(tx)[get_timestamp(time1):get_timestamp(time2)] + shifts[str(n)], color = c, label = 'Thermistor '+n,  linestyle = 'none', marker = '.', markersize = 2)
    avg_value(np.asarray(tx) + shifts[str(n)], get_timestamp(start_avg), get_timestamp(end_avg), n)

#Normalizes the plot of the running average of a thermistor dataset
def norm_mean(tx, c, n):
    mean_plot([x + shifts[str(n)] for x in tx], c, n)

#prints the average value of a thermistor between two points in time (in minutes) and calculates the standard error of this value
def avg_value(tx, start, stop, n):
    mean = sum(tx[start: stop]) / len(tx[start: stop])
    std_err = np.std(tx[start: stop]) / np.sqrt(len(tx[start: stop]))
    print("Thermistor " + str(n) + ": " + str(mean))
    print("      (Standard error: " + str(std_err) + ")")
```

```python
#plots the normalized thermistor values
specific_norm(t1, 'r', '1') #Side 1 Module
specific_norm(t2, 'deeppink', '2') #Housing Side Far
specific_norm(t3, 'yellow', '3') #Housing Side Near
specific_norm(t4, 'orange', '4') #Cooling plate
specific_norm(t5, 'purple', '5') #Housing bottom
specific_norm(t6, 'aqua', '6') #Aluminium plate, centered
specific_norm(t7, 'blue', '7') #Aluminum plate, far right

#Calculates and prints mean and standard error for the ambient temperature readout
mean_amb = sum(amb_temp[get_timestamp(start_avg): get_timestamp(end_avg)]) / len(amb_temp[get_timestamp(start_avg): get_timestamp(end_avg)])
std_err = np.std(amb_temp[get_timestamp(start_avg): get_timestamp(end_avg)]) / np.sqrt(len(amb_temp[get_timestamp(start_avg): get_timestamp(end_avg)]))
print("Ambient Temperature: " + str(mean_amb))
print("      (Standard error: " + str(std_err) + ")")

plt.xlabel('Time / min', fontsize =16)
plt.ylabel('Thermistor temperature / $^\circ$C', fontsize =16)

plt.legend(prop={'size':14}, bbox_to_anchor=(1., 1), markerscale=5)
plt.title("Cu-Module Contact Test, 0.66 W/m.K putty")
plt.grid()
plt.show()
```
