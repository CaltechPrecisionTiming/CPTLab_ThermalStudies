```python
#Import Statements
import numpy as np
import matplotlib.pyplot as plt
import csv
import datetime
import math
from scipy import optimize

#Path of CSV Dataset
path_name = 'CSVDatasets/210709000000_210709090000.csv'

#Arrays for thermistor data
t0,t1,t2,t3,t4,t5,t6,t7 = [],[],[],[],[],[],[],[]

#Time array
time_u = []
```

```python
#Converts ADC values to celsius for thermistors 2-7
def cal(R):
    R = adc2R(float(R))
    beta=3892.0
    T0=25.0+273.15
    R0=10000
    r_inf = R0*math.exp(-beta/T0)
    T = beta/math.log(R/r_inf)
    return T - 273.15

#Converts ADC values to celsius for thermistors 0 and 1
def cal_surface(R):
    R = adc2R(float(R))
    beta=3940
    T0=25.0+273.15
    R0=10000
    r_inf = R0*math.exp(-beta/T0)
    T = beta/math.log(R/r_inf)
    return T - 273.15

#Helper function for both cal conversions
def adc2R(c, R=19e3, Vref=2.5):
    V = c*Vref/4096
    return V*R/(Vref-V);

#Fixes data values in csv file using err and cal
with open(path_name) as csvfile: 
    data = csv.reader(csvfile)
    for i, row in enumerate(data):
        t0.append(cal_surface(row[1]))
        t1.append(cal_surface(row[2]))
        t2.append(cal(row[3]))
        t3.append(cal(row[4]))
        t4.append(cal(row[5]))
        t5.append(cal(row[6]))
        t6.append(cal(row[7]))
        t7.append(cal(row[8]))
        time_u.append(row[13])
    print(i)
    
print(time_u[-3:])
print(t0[-3:])
print(len(time_u))
```

```python
first_time = datetime.datetime.strptime(time_u[0],'%Y-%m-%d %H:%M:%S')

time_s = []
for i in time_u: #converts the hours time stamp to time elapsed
    elapsed = datetime.datetime.strptime(i,'%Y-%m-%d %H:%M:%S') - first_time
    time_s.append((elapsed.total_seconds())/60)

#Allows splicing of the plot at specific timestamps
def get_timestamp(minutes):
    for i in time_s:
        if i > minutes:
            return time_s.index(i) - 1

#Regular Plotting Function for Matplotlib
def raw_plot(tx,c,n):
    #[get_timestamp(280):get_timestamp(340)]
    plt.plot(time_s[get_timestamp(25):get_timestamp(50)], np.asarray(tx)[get_timestamp(25):get_timestamp(50)], color = c, label = 'Thermistor '+n,  linestyle = 'none', marker = '.', markersize = 2)

#Normalizes the graph of one thermistor to match the plot of another by subtracting a fixed value
def norm(tx, t_match, match_time, c, n):
    time_ints = [int(i) for i in time_s]
    time_index = time_ints.index(match_time)
    diff = tx[time_index] - t_match[time_index]
    plt.plot(time_s[get_timestamp(280):get_timestamp(340)], np.asarray(tx)[get_timestamp(280):get_timestamp(340)] - diff, color = c, label = 'Thermistor '+n,  linestyle = 'none', marker = '.', markersize = 2)

#Plots the running average of a thermistor dataset
def mean_plot(tx,c,n):
    total_sum = 0
    mean_tx = []
    for x in enumerate(tx):
        total_sum += x[1]
        mean = total_sum / (x[0]+1)
        mean_tx.append(mean)
    plt.plot(time_s[get_timestamp(280):get_timestamp(340)], np.asarray(mean_tx)[get_timestamp(280):get_timestamp(340)], color = c, label = 'Thermistor '+n,  linestyle = 'none', marker = '.', markersize = 2)
```
